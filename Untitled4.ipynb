```python
import numpy as np
import matplotlib.pyplot as plt

def load_shift_vector(dim):
    """Generate random shift vector in [-80,80]^D"""
    return np.random.uniform(-80, 80, dim)

def create_rotation_matrix(dim, condition_number=2):
    """Create rotation matrix using Gram-Schmidt"""
    M = np.random.randn(dim, dim)
    Q, _ = np.linalg.qr(M)
    return Q

def logistic_map(x0, n_iter, r=3.99):
    """Generate chaotic sequence using logistic map"""
    sequence = [x0]
    for _ in range(n_iter-1):
        x = r * sequence[-1] * (1 - sequence[-1])
        sequence.append(x)
    return np.array(sequence)
```

```python
# ==============================
# CEC Benchmark Functions
# ==============================

# CEC 2014 Functions (F1-F15)
def cec2014_f1(x):
    """Rotated High Conditioned Elliptic Function"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    return sum(1e6**(i/(dim-1)) * z[i]**2 for i in range(dim))

def cec2014_f4(x):
    """Shifted and Rotated Rosenbrock's Function"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, 0.02048*(x - o)) + 1
    return sum(100*(z[i]**2 - z[i+1])**2 + (z[i]-1)**2 for i in range(dim-1))

# CEC 2017 Functions (F16-F25)
def cec2017_f16(x):
    """Hybrid Function 1"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    return sum(z[i]**2 - 10*np.cos(2*np.pi*z[i]) + 10 for i in range(dim))

# CEC 2020 Functions (F26-F35)
def cec2020_f26(x):
    """Composition Function 1"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    return sum(z[i]**4 - 16*z[i]**2 + 5*z[i] for i in range(dim)) / dim

# CEC 2022 Functions (F36-F45)
def cec2022_f36(x):
    """Hybrid Function with Global Op"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    return -20*np.exp(-0.2*np.sqrt(sum(z**2)/dim)) - np.exp(sum(np.cos(2*np.pi*z))/dim) + 20 + np.e
#Hybrid Function 5 (CEC2017 F24)
def cec2017_f24(x):
    """Hybrid Function 5 (Quadratic + Rastrigin + Griewank)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    partitions = np.array_split(z, 3)

    fns = [
        lambda x: np.sum(x**2),
        lambda x: 10*dim + np.sum(x**2 - 10*np.cos(2*np.pi*x)),
        lambda x: 1 + np.sum(x**2/4000) - np.prod(np.cos(x/np.sqrt(1+np.arange(len(x)))))
    ]

    return sum(fn(part) for fn, part in zip(fns, partitions))
#Composition Function 5 (CEC2020 F35)
def cec2020_f35(x):
    """Composition of Rotated Ackley, Rosenbrock, Sphere"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)

    f1 = -20*np.exp(-0.2*np.sqrt(np.mean(z**2))) - np.exp(np.mean(np.cos(2*np.pi*z))) + 20 + np.e
    f2 = sum(100*(z[i]**2 - z[i+1])**2 + (z[i]-1)**2 for i in range(dim-1))
    f3 = np.sum(z**2)

    return 0.4*f1 + 0.4*f2 + 0.2*f3 + 20*np.sin(5*np.pi*np.linalg.norm(z))
#Rotated Hybrid Function 6 (CEC2022 F48)
def cec2022_f48(x):
    """Rotated Hybrid (Weierstrass + HappyCat + HGBat)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim, 100)  # High condition number
    z = np.dot(M, x - o)

    # Weierstrass
    f1 = sum([sum([0.5**k * np.cos(2*np.pi*3**k*(zi + 0.5)) for k in range(20)]) for zi in z])

    # HappyCat
    s = np.sum(z**2)
    f2 = abs(s - dim)**0.25 + (0.5*s + np.sum(z))/dim + 0.5

    # HGBat
    s1, s2 = np.sum(z), np.sum(z**2)
    f3 = np.sqrt(abs(s2**2 - s1**2)) + (0.5*s2 + s1)/dim + 0.5

    return f1 + f2 + f3 + 10*np.random.rand()  # Stochastic component

def rotated_expanded_scaffer(x):
    """F31: Rotated Expanded Scaffer's Function"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, 0.1*(x - o))
    return sum(0.5 + (np.sin(np.sqrt(z[i]**2 + z[(i+1)%dim]**2))**2 - 0.5) /
              (1 + 0.001*(z[i]**2 + z[(i+1)%dim]**2))**2 for i in range(dim))

def hybrid_ackley_rastrigin(x):
    """F32: Hybrid Function 7 (Ackley + Rastrigin)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    partitions = np.array_split(z, 2)
    fns = [
        lambda x: -20*np.exp(-0.2*np.sqrt(np.mean(x**2))) - np.exp(np.mean(np.cos(2*np.pi*x))) + 20 + np.e,
        lambda x: 10*len(x) + np.sum(x**2 - 10*np.cos(2*np.pi*x))
    ]
    return sum(fn(part) for fn, part in zip(fns, partitions))

def composition_rosenbrock_schwefel(x):
    """F33: Composition Function 7 (Rosenbrock + Schwefel)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, 0.02048*(x - o))
    return 0.6*sum(100*(z[i]**2 - z[i+1])**2 + (z[i]-1)**2 for i in range(dim-1)) + \
           0.4*(418.9829*dim - np.sum(z * np.sin(np.sqrt(np.abs(z)))))

def shifted_rotated_katsuura(x):
    """F34: Shifted and Rotated Katsuura Function"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, 0.05*(x - o))
    product = 1.0
    for i in range(dim):
        product *= 1 + (i+1)*sum([abs(2**j * z[i] - round(2**j * z[i]))/2**j for j in range(1,32)])
    return 10/dim**2 * product - 10/dim**2 + 10*np.sum(z**2)

def hybrid_griewank_weierstrass(x):
    """F35: Hybrid Function 8 (Griewank + Weierstrass)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    partitions = np.array_split(z, 2)
    fns = [
        lambda x: 1 + np.sum(x**2/4000) - np.prod(np.cos(x/np.sqrt(1+np.arange(len(x))))),
        lambda x: sum(sum(0.5**k * np.cos(2*np.pi*3**k*(xi + 0.5)) for k in range(20)) for xi in x)
    ]
    return sum(fn(part) for fn, part in zip(fns, partitions))

def composition_hgbat_happycat(x):
    """F36: Composition Function 8 (HGBat + HappyCat)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    s1, s2 = np.sum(z), np.sum(z**2)
    return 0.5*(np.sqrt(abs(s2**2 - s1**2)) + (0.5*s2 + s1)/dim + 0.5) + \
           0.5*(abs(s2 - dim)**0.25 + (0.5*s2 + s1)/dim + 0.5)

def rotated_modified_schwefel(x):
    """F37: Rotated Modified Schwefel's Function"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim, 1000)
    z = np.dot(M, 10*(x - o))
    return 418.9829*dim - np.sum(z * np.sin(np.sqrt(np.abs(z)))) + 20*(1 - np.cos(np.pi*np.linalg.norm(z)/dim))

def hybrid_rastrigin_happycat(x):
    """F38: Hybrid Function 9 (Rastrigin + HappyCat)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    partitions = np.array_split(z, 2)
    fns = [
        lambda x: 10*len(x) + np.sum(x**2 - 10*np.cos(2*np.pi*x)),
        lambda x: abs(np.sum(x**2) - len(x))**0.25 + (0.5*np.sum(x**2) + np.sum(x))/len(x) + 0.5
    ]
    return sum(fn(part) for fn, part in zip(fns, partitions))

def composition_ackley_griewank(x):
    """F39: Composition Function 9 (Ackley + Griewank)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    return 0.7*(-20*np.exp(-0.2*np.sqrt(np.mean(z**2))) - np.exp(np.mean(np.cos(2*np.pi*z))) + 20 + np.e) + \
           0.3*(1 + np.sum(z**2/4000) - np.prod(np.cos(z/np.sqrt(1+np.arange(dim)))))

def shifted_rotated_expanded_griewank(x):
    """F40: Shifted and Rotated Expanded Griewank's Function"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, 600*(x - o))
    return 1 + np.sum(z**2/4000) - np.prod(np.cos(z/np.sqrt(1+np.arange(dim)))) + 0.1*np.sum(np.abs(z))

def hybrid_schwefel_sphere(x):
    """F41: Hybrid Function 10 (Schwefel + Sphere)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    partitions = np.array_split(z, 2)
    fns = [
        lambda x: 418.9829*len(x) - np.sum(x * np.sin(np.sqrt(np.abs(x)))),
        lambda x: np.sum(x**2)
    ]
    return sum(fn(part) for fn, part in zip(fns, partitions))

def composition_rastrigin_katsuura(x):
    """F42: Composition Function 10 (Rastrigin + Katsuura)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    return 0.6*(10*dim + np.sum(z**2 - 10*np.cos(2*np.pi*z))) + \
           0.4*(10/dim**2 * np.prod([1 + (i+1)*sum([abs(2**j * zi - round(2**j * zi))/2**j for j in range(1,32)]) for i,zi in enumerate(z)]))

def rotated_lunacek(x):
    """F43: Rotated Lunacek Function"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    s = 1 - 1/(2*np.sqrt(dim+20)-8.2)
    return np.minimum(np.sum((z - 2.5)**2), np.sum((z + 2.5)**2)) + 10*(dim - np.sum(np.cos(2*np.pi*(z - 2.5))))

def hybrid_rosenbrock_hgbat(x):
    """F44: Hybrid Function 11 (Rosenbrock + HGBat)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    partitions = np.array_split(z, 2)
    fns = [
        lambda x: sum(100*(x[i]**2 - x[i+1])**2 + (x[i]-1)**2 for i in range(len(x)-1)),
        lambda x: np.sqrt(abs(np.sum(x**2)**2 - np.sum(x)**2)) + (0.5*np.sum(x**2) + np.sum(x))/len(x) + 0.5
    ]
    return sum(fn(part) for fn, part in zip(fns, partitions))

def composition_weierstrass_griewank(x):
    """F45: Composition Function 11 (Weierstrass + Griewank)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    return 0.5*sum(sum(0.5**k * np.cos(2*np.pi*3**k*(zi + 0.5)) for k in range(20)) for zi in z) + \
           0.5*(1 + np.sum(z**2/4000) - np.prod(np.cos(z/np.sqrt(1+np.arange(dim)))))

def shifted_rotated_expanded_rosenbrock(x):
    """F46: Shifted and Rotated Expanded Rosenbrock's Function"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, 0.02048*(x - o)) + 2.048/2
    return sum(100*(z[i]**2 - z[i+1])**2 + (z[i]-1)**2 for i in range(dim-1)) + 10*np.sum(np.abs(z))

def hybrid_ackley_schwefel(x):
    """F47: Hybrid Function 12 (Ackley + Schwefel)"""
    dim = len(x)
    o = load_shift_vector(dim)
    M = create_rotation_matrix(dim)
    z = np.dot(M, x - o)
    partitions = np.array_split(z, 2)
    fns = [
        lambda x: -20*np.exp(-0.2*np.sqrt(np.mean(x**2))) - np.exp(np.mean(np.cos(2*np.pi*x))) + 20 + np.e,
        lambda x: 418.9829*len(x) - np.sum(x * np.sin(np.sqrt(np.abs(x))))
    ]
    return sum(fn(part) for fn, part in zip(fns, partitions))

```

```python
# ==============================
# Algorithm Implementations
# ==============================
def chaotic_mwso(obj_func, dim, bounds, max_iter=1000, n_sharks=30):
    """Hybrid Chaotic-MWSO"""
    lb, ub = bounds
    sharks = np.random.uniform(lb, ub, (n_sharks, dim))
    fitness = np.array([obj_func(ind) for ind in sharks])
    best_idx = np.argmin(fitness)
    gbest = sharks[best_idx].copy()
    f_best = fitness[best_idx]

    # Chaotic sequence initialization
    chaos = logistic_map(np.random.rand(), max_iter)

    convergence = []

    for t in range(max_iter):
        w = 0.9 - 0.5*t/max_iter  # Inertia weight decay
        c = chaos[t]  # Chaotic parameter

        for i in range(n_sharks):
            # Chaotic velocity update
            r1, r2 = np.random.rand(2)
            A = 2 * c - 1
            velocity = w * sharks[i] + A * r1 * (gbest - sharks[i])

            # Quasi-oppositional learning
            if np.random.rand() < 0.5*c:
                candidate = lb + ub - sharks[i] + c*(ub - lb)*0.1
            else:
                candidate = sharks[i] + velocity

            candidate = np.clip(candidate, lb, ub)
            f_candidate = obj_func(candidate)

            if f_candidate < fitness[i]:
                sharks[i] = candidate
                fitness[i] = f_candidate
                if f_candidate < f_best:
                    gbest = candidate.copy()
                    f_best = f_candidate

        convergence.append(f_best)

    return gbest, f_best, convergence

def standard_mwso(obj_func, dim, bounds, max_iter=1000, n_sharks=30):
    """Core MWSO Implementation"""
    lb, ub = bounds
    sharks = np.random.uniform(lb, ub, (n_sharks, dim))
    fitness = np.array([obj_func(ind) for ind in sharks])
    best_idx = np.argmin(fitness)
    gbest = sharks[best_idx].copy()
    f_best = fitness[best_idx]

    convergence = []

    for t in range(max_iter):
        w = 0.9 - 0.5*t/max_iter

        for i in range(n_sharks):
            r1, r2 = np.random.rand(2)
            A = 2 * r1 - 1
            velocity = w * sharks[i] + A * r2 * (gbest - sharks[i])

            candidate = sharks[i] + velocity
            candidate = np.clip(candidate, lb, ub)
            f_candidate = obj_func(candidate)

            if f_candidate < fitness[i]:
                sharks[i] = candidate
                fitness[i] = f_candidate
                if f_candidate < f_best:
                    gbest = candidate.copy()
                    f_best = f_candidate

        convergence.append(f_best)

    return gbest, f_best, convergence

# Basic Random Search
def random_search(obj_func, dim, bounds, max_iter=1000):
    """Pure exploration with no exploitation"""
    lb, ub = bounds
    best_solution = np.random.uniform(lb, ub, dim)
    best_fitness = obj_func(best_solution)
    convergence = [best_fitness]

    for _ in range(max_iter-1):
        candidate = np.random.uniform(lb, ub, dim)
        current_fitness = obj_func(candidate)
        if current_fitness < best_fitness:
            best_solution = candidate.copy()
            best_fitness = current_fitness
        convergence.append(best_fitness)

    return best_solution, best_fitness, convergence

#Hill climbing
def hill_climbing(obj_func, dim, bounds, max_iter=1000):
    """Local search prone to stagnation"""
    lb, ub = bounds
    current = np.random.uniform(lb, ub, dim)
    current_fit = obj_func(current)
    convergence = [current_fit]

    for _ in range(max_iter-1):
        neighbor = current + np.random.normal(0, 0.1, dim)
        neighbor = np.clip(neighbor, lb, ub)
        neighbor_fit = obj_func(neighbor)

        if neighbor_fit < current_fit:
            current = neighbor.copy()
            current_fit = neighbor_fit

        convergence.append(current_fit)

    return current, current_fit, convergence

#Basic Genetic Algorithm
def simple_ga(obj_func, dim, bounds, max_iter=1000, pop_size=30):
    """No elite preservation + random crossover"""
    lb, ub = bounds
    pop = np.random.uniform(lb, ub, (pop_size, dim))
    fitness = np.array([obj_func(ind) for ind in pop])
    best_idx = np.argmin(fitness)
    best = pop[best_idx].copy()
    best_fit = fitness[best_idx]
    convergence = [best_fit]

    for _ in range(max_iter-1):
        # Tournament selection
        parents = pop[np.random.choice(pop_size, 2, replace=False)]
        # Single-point crossover
        crossover_pt = np.random.randint(1, dim)
        child = np.concatenate([parents[0][:crossover_pt],
                               parents[1][crossover_pt:]])
        # Random mutation
        child += np.random.normal(0, 0.1, dim)
        child = np.clip(child, lb, ub)

        # Replace worst member
        worst_idx = np.argmax(fitness)
        child_fit = obj_func(child)
        if child_fit < fitness[worst_idx]:
            pop[worst_idx] = child
            fitness[worst_idx] = child_fit
            if child_fit < best_fit:
                best = child.copy()
                best_fit = child_fit

        convergence.append(best_fit)

    return best, best_fit, convergence

# Basic Simulated Annealing
def simulated_annealing(obj_func, dim, bounds, max_iter=1000):
    """Fixed cooling schedule"""
    lb, ub = bounds
    current = np.random.uniform(lb, ub, dim)
    current_fit = obj_func(current)
    T = 1.0
    cooling = 0.95
    convergence = [current_fit]

    for _ in range(max_iter-1):
        neighbor = current + np.random.normal(0, T, dim)
        neighbor = np.clip(neighbor, lb, ub)
        neighbor_fit = obj_func(neighbor)

        if neighbor_fit < current_fit or \
           np.random.rand() < np.exp(-(neighbor_fit - current_fit)/T):
            current = neighbor.copy()
            current_fit = neighbor_fit

        T *= cooling
        convergence.append(current_fit)

    return current, current_fit, convergence
#Blind Algorithm
def blind_guided_search(obj_func, dim, bounds, max_iter=1000):
    """Single-solution search with fixed-step random walks"""
    lb, ub = bounds
    current = np.random.uniform(lb, ub, dim)
    current_fit = obj_func(current)
    convergence = [current_fit]
    step_size = 0.01 * (ub - lb)  # Fixed small step

    for _ in range(max_iter-1):
        direction = np.random.choice([-1, 1], dim)
        candidate = current + direction * step_size
        candidate = np.clip(candidate, lb, ub)
        candidate_fit = obj_func(candidate)

        if candidate_fit < current_fit:
            current = candidate.copy()
            current_fit = candidate_fit

        convergence.append(current_fit)

    return current, current_fit, convergence
```
